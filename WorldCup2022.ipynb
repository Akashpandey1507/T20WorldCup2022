# Which Venues hosted most of the matches.

```python
plt.figure(figsize=(25,10))
plt.title('Venues for hosted matches', fontsize= 25)
plt.xlabel('', fontsize= 20)
plt.ylabel('', fontsize= 20)

count_plot = sns.countplot(data=df, x='Venue')

for p in count_plot.patches:
    count_plot.annotate(format(p.get_height(), '.0f'), (p.get_x() + p.get_width() / 2., p.get_height()), 
    ha = 'center', va = 'center', xytext = (0, 10), textcoords = 'offset points', fontsize=12)

plt.show()
```

# Result:
* SCG and MCG were hosted most of matches in T20 World Cup 2022.

# Who is the highest match winner from group stage to finale.


```python
plt.figure(figsize=(25,10))
plt.title('Most Matches Win', fontsize= 25)
plt.xlabel('', fontsize= 20)
plt.ylabel('', fontsize= 20)

count_plot = sns.countplot(data=df, x='Winner')

for p in count_plot.patches:
    count_plot.annotate(format(p.get_height(), '.0f'), (p.get_x() + p.get_width() / 2., p.get_height()), 
    ha = 'center', va = 'center', xytext = (0, 10), textcoords = 'offset points', fontsize=12)

plt.show()
```

# Result 
* England is the most match winner in 2022 T20 World Cup with 5 matches.
* India and Pakistan are 2nd most match winner in T20 World Cup 2022 with 4 and 4 matches respectively.

# Who is the most toss winner and result of the toss

```python
plt.figure(figsize=(25,10))
plt.title('Toss win by team', fontsize= 25)
plt.xlabel('', fontsize= 20)
plt.ylabel('', fontsize= 20)

count_plot = sns.countplot(data=df, x='Toss Winner',hue='Toss Decision')

for p in count_plot.patches:
    count_plot.annotate(format(p.get_height(), '.0f'), (p.get_x() + p.get_width() / 2., p.get_height()), 
    ha = 'center', va = 'center', xytext = (0, 10), textcoords = 'offset points', fontsize=12)

plt.show()
```

# Result
* England have won the most of the toss with bat first.

# Who is the Top Sccorer in T20 World Cup

```python
Top = df.groupby('Top Scorer')['Highest Score'].sum().reset_index()
Top
```

```python

plt.figure(figsize=(25, 10))
plt.title('Highest Run by Individual', fontsize=25)
plt.xlabel('', fontsize=20)
plt.ylabel('', fontsize=20)
plt.xticks(rotation=90, fontsize=15)


line_plot = sns.lineplot(data=Top, x='Top Scorer', y='Highest Score', marker='o', markersize=8
,color='g')

# Remove x-axis and y-axis tick labels

line_plot.set_yticklabels([])

# Add count labels on each data point
for line in line_plot.lines:
    for x, y in zip(line.get_xdata(), line.get_ydata()):
        line_plot.annotate(format(y, '.0f'), (x, y), ha='center', va='bottom', fontsize=12)

plt.show()
```

# Result:
* The King Virat Kohali is most highet runner in T20 World Cup 
* Surya Kumar Yadav is second most highet runner in T20 World Cup
but Unfortunetly Indian did not played finale.

```python

plt.figure(figsize=(25, 10))
plt.title('Most win Player Of The Match', fontsize=25)
plt.xlabel('', fontsize=20)
plt.ylabel('', fontsize=20)
plt.xticks(rotation=90)

count_plot = sns.countplot(data=df, x='Player Of The Match')

# Remove y-axis tick labels
count_plot.set_yticklabels([])

for p in count_plot.patches:
    count_plot.annotate(format(p.get_height(), '.0f'), (p.get_x() + p.get_width() / 2., p.get_height()), 
    ha='center', va='center', xytext=(0, 10), textcoords='offset points', fontsize=12)

plt.show()
```

# Who is the best Bowler and best bowler figure

```python
# Create a figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(15, 7))

# Plot 1: Bar plot
count_plot1 = sns.countplot(data=df, x='Best Bowler', ax=axes[0])
axes[0].set_title('Best Bowler', fontsize=25)
axes[0].set_xlabel('', fontsize=20)
axes[0].set_ylabel('', fontsize=20)
axes[0].tick_params(axis='x', rotation=90)  # Set rotation in xticks for subplot 1
axes[0].set_yticklabels([])  # Remove y-axis tick labels for subplot 1

# Add count labels on each bar in subplot 1
for p in count_plot1.patches:
    count_plot1.annotate(format(p.get_height(), '.0f'), (p.get_x() + p.get_width() / 2., p.get_height()), 
                         ha='center', va='center', xytext=(0, 10), textcoords='offset points', fontsize=12)

# Plot 2: Bar plot
count_plot2 = sns.countplot(data=df, x='Best Bowling Figure', ax=axes[1])
axes[1].set_title('Best Bowling Figure', fontsize=25)
axes[1].set_xlabel('', fontsize=20)
axes[1].set_ylabel('', fontsize=20)
axes[1].tick_params(axis='x', rotation=90)  # Set rotation in xticks for subplot 2
axes[1].set_yticklabels([])  # Remove y-axis tick labels for subplot 2

# Add count labels on each bar in subplot 2
for p in count_plot2.patches:
    count_plot2.annotate(format(p.get_height(), '.0f'), (p.get_x() + p.get_width() / 2., p.get_height()), 
                         ha='center', va='center', xytext=(0, 10), textcoords='offset points', fontsize=12)

# Adjust the layout spacing between subplots
plt.tight_layout()

# Show the plots
plt.show()

```

# Result
* Sam curran is the best bowler in T20 World Cup 2022.

# What the best venue for first and second inning score.

```python
# Create a figure with a single plot
plt.figure(figsize=(25, 10))

# Plotting the first inning scores
sns.barplot(data=df, x='Venue', y='First Innings Score')
plt.title('First Inning Scores by Venue', fontsize=20)
plt.xlabel('Venue', fontsize=15)
plt.ylabel('First Inning Score', fontsize=15)

plt.gca().set_yticklabels([])  # Remove y-axis tick labels

# Add count labels on each bar
for p in plt.gca().patches:
    plt.gca().annotate(format(p.get_height(), '.0f'), (p.get_x() + p.get_width() / 2., p.get_height()),
                       ha='center', va='center', xytext=(0, 10), textcoords='offset points', fontsize=12)

# Show the first plot
plt.show()
```

# Result:
* Best Venue for first in SCG .

```python

# Create a new figure for the second plot
plt.figure(figsize=(25, 10))

# Plotting the second inning scores
sns.barplot(data=df, x='Venue', y='Second Innings Score')
plt.title('Second Inning Scores by Venue', fontsize=20)
plt.xlabel('Venue', fontsize=15)
plt.ylabel('Second Inning Score', fontsize=15)

plt.gca().set_yticklabels([])  # Remove y-axis tick labels

# Add count labels on each bar
for p in plt.gca().patches:
    plt.gca().annotate(format(p.get_height(), '.0f'), (p.get_x() + p.get_width() / 2., p.get_height()),
                       ha='center', va='center', xytext=(0, 10), textcoords='offset points', fontsize=12)

# Show the second plot
plt.show()
```

# Result:
* Best venue for Second is The Gabba and Adelaide Oval
